#### Two Pointer or Sliding Window

Max points you can obtain from cards give k = 4

you can either go from front or back but you have to pick 4 cards which maximizes your output

```python
nums = [6,2,3,4,7,2,1,7,1]
k = 4
lsum =  maxsum = 0
end = len(nums) - 1
for i in range(0,k):
    lsum += nums[i]
    maxsum = lsum
for y in range(k-1,0,-1):
    lsum -= nums[y]
    lsum += nums[end]
    end -= 1
    maxsum = max(maxsum, lsum)
print(maxsum)
```
